```python
import numpy as np
nums=np.arange(5)
nums
```

```python
#plt.figure()
#plt.plot(nums,nums)
plt.figure()
plt.style.use('fivethirtyeight')
plt.plot(nums,nums*nums,label="sub1")
plt.plot(nums,nums**3,label="sub2")
plt.legend(frameon=True,facecolor="white")
plt.title("Chart Title")
plt.xlabel("Year")
plt.xticks([0,1,2,3,4])
#plt.xlim(1,4)
```
